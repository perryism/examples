```python
from bokeh.io import output_notebook
import pandas as pd
output_notebook()
from bokeh.charts import Donut, show
```

```python
p = [ 0.2424,  0.4069,  0.1041,  0.1135,  0.1331]
data = pd.Series(p, index = ['apple', 'orange', 'banana', 'kiwi', 'watermelon'])
pie_chart = Donut(data)
show(pie_chart)
```
