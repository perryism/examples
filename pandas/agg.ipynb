```python
import pandas as pd

df = pd.DataFrame([
    ['Perry', 150], 
    ['Veronica', 16], 
    ['Perry', 150], 
    ['Perry', 150], 
    ['Veronica', 34], 
    ['Veronica', 12], 
    ['Zachary', 3]], 
    columns=['name', 'spending'])
df
```

```python
aggregation = {
    'spending': {
        'total_spending': 'sum',
        'avg_spending': 'mean'
    }
}

df.groupby('name').agg(aggregation)
```
